```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

from pandas import read_csv
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARIMA
from math import sqrt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
data = sio.loadmat('sample_west.mat')
wt_data = data['t_data']
w_data = data['y_data']
wt_ans = data['t_ans']
w_ans = data['y_ans']

print(w_data.shape)

plot_acf(w_data)
plot_pacf(w_data)
plt.show()

plt.plot(w_data)
plt.show()
```

```python
sm.tsa.adfuller(w_data)
```

```python
w_pi = []
n = len(w_data)
m = int(n/10)
for i in range(10):
    data = w_data[:m]
    w = ARIMA(data, order=[1,0,0])
    w_fit = w.fit()
    wvalue = w_fit.arparams
    w_pi.append(wvalue)
    w_data = w_data[m:]
```
